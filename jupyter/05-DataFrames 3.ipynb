## Retrieve Rows by Index Label with `.loc[]`

#### 检索row 

- 按index名： dataframe.loc[index_label]
  - index要输入具体label,而不是index位置
  - 注意，index在pandas dataframe中不一定要unqiue,所以一个index有可能检索出多行
  - 按照index 来选/retrieve行，index可为index series
  - dataframe.loc["a" : "z"],注意两边都是包含的
  - 注意，如果输入的index本不存在，ta依旧会返回一行全部为nan的数据，一定要小心
- 按行位置： dataframe.iloc[index_position]
    - 输入的是行的位置/index position,从0起。可以是一个index series
- 两者都可以用： dataframe.ix[]
    - bond.ix[["Diamonds are Forever", "Moonraker", "Spectre"]]
    - bond.iloc[[14, 5, 3, 2]]

#### 检索某行某列的值

- loc， iloc， ix 后面如果只写一项/一个list，就是指行
- 若后面是[list1, list2]
    - list1 就是index/行名
    - list2就是列名/column index
- loc还是指label/名称检索
- iloc 还是数字/index 位置/第几个
- ix可以混合

    bond.loc["Moonraker", ["Actor", "Budget", "Year"]]
    bond.iloc[14, [5, 3, 2]]
    bond.ix[20, "Budget"]

## Rename Index Labels or Columns in a `DataFrame`

- 重命名列名-1： dataframe.rename（columns = dictionary)
    - dictionary的key：原本列名，value为新列名
    - bond.rename(columns = {"Year" : "Release Date", "Box Office" : "Revenue"}, inplace = True)
- 重命名列名-2：dataframe.columns
    - 直接修改attribute就能修改所有name，如下（这样比较繁琐）
    - 如，bond.columns = ["Year of Release", "Actor", "Director", "Gross", "Cost", "Salary"]
- 重命名index名:dataframe.rename(index = dictionary)
    bond.rename(index = {"Dr. No" : "Doctor No", 
    "GoldenEye" : "Golden Eye",
    "The World Is Not Enough" : "Best Bond Movie Ever"}, inplace = True)
- 删除某些列某些行：dataframe.drop(label_list, index,inplace)
    - Index=0/"rows"是处理行，=1/"columns"处理列
    - 对应index指明的行列，label_list就是label series,会删除这些行列
    - 如 bond.drop(labels = ["Box Office", "Bond Actor Salary", "Actor"], axis = "columns", inplace = True)
- 直接删除某一列： datarame.pop(columns)
    - 这样会直接在原dataframe上处理，(删除某一个series)删除label,并且返回
    - 如 actor = bond.pop("Actor")
- 从dataframe删除某些series
    - del bond["Director"]

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
bond.rename(columns = {"Year" : "Release Date", "Box Office" : "Revenue"}, inplace = True)
```

## Create Random Sample

- dataframe.sample(n,axis="columns"/"index") 
    - 随机抽样几行/几列，前后顺序也会被打乱
    - 若不想指定具体是几行，可以用frac参数，在总行数/列数中取某百分比个行/列
    - o:index, 1: column
    - bond.sample(n = 3, axis = "columns")
    - bond.sample(frac = 0.25, axis = "index")
- dataframe..nsmallest(n,columns)    dataframe.nlargest()
    - 取整个集合的最大/最小n行，这个最大最小由column参数指明
    - 也可以Series.nsmallest()
    - bond.nsmallest(n = 2, columns = "Box Office")
    - bond["Year"].nsmallest(2)
- dataframe.where(condition)
    - 一种retrieve方法，不会返回subset,而是返回全集，但是没被选中的会被弄成NaN
    - mask = bond["Actor"] == "Sean Connery"
    - bond.where(mask)
- dataframe.query（'Col_name == "A"'）
    - 很快
    - 只能处理没有空格的column name
    - 返回符合的所有行，且直接打列名就可以识别
    - bond.query("Actor in ['Timothy Dalton', 'George Lazenby']")
    - bond.query("Actor not in ['Timothy Dalton', 'George Lazenby']")
    - bond.query("Box_Office > 600")

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
bond.sample(n = 3, axis = "columns").head()
```

## A Review of the `.apply()` Method on Single Columns

- dataframe[col].apply(function) 单列的apply处理
    - 将某series/列所有的item进行function处理，return所有y形成的新series/列
    - bond[col] = bond[col].apply(convert_to_string_and_add_millions)
    - 其实也是按行处理，但是此时每一行只有一个value
- dataframe.apply(func, axis = "columns")
    - dataframe中的每一行作为input，也就是以series作为input，每一行按照function，产出一个y,return所有y形成的新series/列
    - 可以实现多列间的组合计算

    def good_movie(row):

        actor = row[1]
        budget = row[4]

        if actor == "Pierce Brosnan":
            return "The best"
        elif actor == "Roger Moore" and budget > 40:
            return "Enjoyable"
        else:
            return "I have no clue"

    bond.apply(good_movie, axis = "columns")

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def convert_to_string_and_add_millions(number):
    return str(number) + " MILLIONS!"
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    bond[col] = bond[col].apply(convert_to_string_and_add_millions)
```

```python
bond.head(3)
```

## The `.apply()` Method with Row Values

```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
def good_movie(row):
    
    actor = row[1]
    budget = row[4]
    
    if actor == "Pierce Brosnan":
        return "The best"
    elif actor == "Roger Moore" and budget > 40:
        return "Enjoyable"
    else:
        return "I have no clue"
    
bond.apply(good_movie, axis = "columns")
```

## The `.copy()` Method

- directors = bond["Director"]
    - 此时，若修改directors,会直接改变bond
    - 如果将dataframe 中截取的一部分存入到另外一个变量，那么新的变量其实是引用了原dataframe,换句话说，就是新的变量代表了原本dataframe中的一部分，这一部分进行改变的时候，原dataframe也会随之变
    - 所以不希望改变新变量就影响原dataframe的话，就要用到dataframe.copy()
- directors = bond["Director"].copy()
    - 此时，修改directors就不会改变bond了


```python
bond = pd.read_csv("jamesbond.csv", index_col = "Film")
bond.sort_index(inplace = True)
bond.head(3)
```

```python
directors = bond["Director"]
directors.head(3)
```

```python
directors["A View to a Kill"] = "Mister John Glen"
```

```python
directors.head(3)
```

```python
bond.head(3)
```

```python
directors = bond["Director"].copy()
directors.head(3)
```

```python
directors["A View to a Kill"] = "Mister John Glen"
```

```python
directors.head(3)
```
