## Filter A `DataFrame` Based On A Condition

#### 单一条件筛选

- df["Gender"]=="Male" 返回T/F
- df[df["Gender"]=="Male"] 选出T的/符合条件的所有row
    - 如 df[df["Salary"]>1000]
    - 如用mark
        - mark=df["Team"]=="Finance"
        - df[mark]
    - 如df[df["Start Date"]<="1985-01-01"]

#### 多条件筛选

- 与 &

    mask1 = df["Gender"] == "Male"
    mask2 = df["Team"] == "Marketing"
    df[mask1 & mask2]

- 或 | 

    df[(mask1 & mask2) | mask3]
    df[mask1 | mask2]

- Series.isin(list) 某一项是否属于后面list中的一个

    mask1 = df["Team"] == "Legal"
    mask2 = df["Team"] == "Sales"
    mask3 = df["Team"] == "Product"
    df[mask1 | mask2 | mask3]

    mark =df["Team"].isin(["Legal","Product","Marketing"])
    df[mark]

- Series.isnull() 返回是否为null的T/F序列
- Series..notnull() 同理可作为Filter条件

    mask = df["Team"].isnull()
    df[mask]

    condition = df["Gender"].notnull()
    df[condition]

- Series.between(a.b) 返回T/F序列

    df[df["Salary"].between(60000, 70000)] #包含两个边界
    df[df["Last Login Time"].between("08:30AM", "12:00PM")]
    df[df["Start Date"].between("1991-01-01", "1992-01-01")]

- Series.duplicated(keep=)
    - keep = "first"，默认,第一条有重复的不算重复/False，之后后面与前面相同才叫duplicate/True(可以先排好序)
    - keep = “last”，设置最后一条不算重复/False，前面所有的算重复
    - keep = False, 设置 只要是有重复，全部标记为True

    mask = df["First Name"].duplicated(keep = False)
    mask = -df["First Name"].duplicated(keep = False) #“-”号表示颠倒T/F
    df[mask]

- dataframe.drop_duplicates(subset = list , keep，inplace)
    - subset是column list,告诉查哪些列是否有重复
    - keep是判断重复的方法
        - first第一条不算，后面算
        - last最后一条不算，前面算
        - False 全部算重复

    df.drop_duplicates(subset = ["First Name"], keep = False)

- Series.unique() 返回数列的Unique列表，nan算为一项
- Series.nunique() 返回有几个Unique值，一般包括nan

    df["Team"].nunique(dropna = False)

## Filter with More than One Condition (OR)

```python
df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.head(3)
```

```python
mask1 = df["Senior Management"]
mask2 = df["Start Date"] < "1990-01-01"

df[mask1 | mask2]
```

```python
mask1 = df["First Name"] == "Robert"
mask2 = df["Team"] == "Client Services"
mask3 = df["Start Date"] > "2016-06-01"

df[(mask1 & mask2) | mask3]
```

## The `.duplicated()` Method

```python
df = pd.read_csv("employees.csv", parse_dates = ["Start Date", "Last Login Time"])
df["Senior Management"] = df["Senior Management"].astype("bool")
df["Gender"] = df["Gender"].astype("category")
df.sort_values("First Name", inplace = True)
df.head(3)
```

```python
mask = ~df["First Name"].duplicated(keep = False)
df[mask]
```
